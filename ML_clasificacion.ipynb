```python
df2=df.copy()
df2.sample(10)
```

```python
df2.drop(columns=['EmployeeID','EmployeeCount','Over18','StandardHours'],axis=1, inplace=True) #columnas no necesarias
df2['BusinessTravel']=df2['BusinessTravel'].astype('category') #columnas categoricas
df2['Department']=df2['Department'].astype('category')
df2['EducationField']=df2['EducationField'].astype('category')
df2['Gender']=df2['Gender'].astype('category')
df2['NumCompaniesWorked']=df2['NumCompaniesWorked'].fillna(0).astype(np.int64) #a columna de entero
df2.info()
```

```python
df2['BusinessTravel'].value_counts()
```

```python
df2['Attrition'].replace({'Yes':1, 'No':0}, inplace=True) #forma manual de pasar a categoria numerica
```

```python
#forma automatica
df2.Gender=df2.Gender.astype('category').cat.codes
df2.EducationField=df2.EducationField.astype('category').cat.codes
df2.BusinessTravel=df2.BusinessTravel.astype('category').cat.codes
df2.Department=df2.Department.astype('category').cat.codes
df2.MaritalStatus=df2.MaritalStatus.astype('category').cat.codes
df2.JobRole=df2.JobRole.astype('category').cat.codes
df2.sample(10)
```

```python
df2.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() #estandarizar MonthlyIncome
df2['MonthlyIncome_scale']=StandardScaler().fit_transform(df2[['MonthlyIncome']])
#df2.drop(columns=['MonthlyIncome'], axis=1, inplace=True) #eliminamos la columna original
df2.sample(10)
```

```python
df2.nunique()
```

```python
#inputacion de datos nulos con la media
mean_TotalWorkingYears=df2['TotalWorkingYears'].mean()
df2.fillna({'TotalWorkingYears' : mean_TotalWorkingYears}, inplace=True)
```

```python
cols_num=df2.columns[[np.issubdtype(dt, np.number) for dt in df2.dtypes]]
fig, axes= plt.subplots(nrows=5, ncols=5, figsize=(18,16))
for i, column in enumerate(cols_num):
  sns.histplot(df2[column], ax=axes[i//5, i%5], kde=False)
```

```python
from sklearn.model_selection import train_test_split
X=df2.drop(columns=['Attrition'], axis=1)
y=df2['Attrition']
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X.head(5)
```

```python
y.head(10)
```

```python
df2['Attrition'].value_counts(normalize=True) #no esta proporcional
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(solver='liblinear', class_weight='balanced')
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
y_pred[0:200]
```

```python
z=X_test.copy()
z['Attrition real']=y_test
z['Attrition predict']=y_pred
z[['Attrition real','Attrition predict']].head(20)
```

```python
coef_df=pd.DataFrame(model.coef_, columns=X.columns) #coeficientes de la ecuacion
#incluir el intercept
coef_df['Intercept']=model.intercept_
coef_df.T
```

```python
#Matriz de confusion
cm=metrics.confusion_matrix(y_test, y_pred, labels=[1,0])
cm_df=pd.DataFrame(cm, index=[i for i in ['1','0']], columns=[i for i in ['predict 1','predict 0']])
plt.figure(figsize=(7,4))
sns.heatmap(cm_df, annot=True, cmap='Accent', fmt='d').set_title('Matriz de confusion')
```

```python
TP=cm_df.iloc[0,0] #true positive
FP=cm_df.iloc[0,1] #true positive
FN=cm_df.iloc[1,0] #false negative
TN=cm_df.iloc[1,1] #true negative
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)
model_score= accuracy_score(y_test, y_pred)
recall_score= recall_score(y_test, y_pred)
precision_score= precision_score(y_test, y_pred)
f1_score= f1_score(y_test, y_pred)

print('Medidas del modelo')
print('----------------------')
print('Acurracy = {0:0.2f}'.format(model_score)) #num de prediciones correctas
print('Recall= {0:0.2f}'.format(recall_score))  #proporcion de predicciones positivas reales correctas
print('Precision= {0:0.2f}'.format(precision_score)) #proporcion de predicciones positivas correctas
print('F1= {0:0.2f}'.format(f1_score)) #media armonica de recall y presicion
print('Roc auc score= {0:0.2f}'.format(roc_auc_score(y_test, y_pred))) #curva que muestra que tan bueno es el modelo
```
