```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_numerico = df.select_dtypes(include='number') #tomar en cuenta solo valores numericos
plt.figure(figsize=(12,6))
df_corr= corr= df_numerico.corr()
df_lt=df_corr.where(np.tril(np.ones(corr.shape)).astype(np.bool))
hmap= sns.heatmap(df_lt, cmap="YlGnBu", annot=True)
```

```python
col_num=['age',	'duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
```

```python
fig, axes=plt.subplots(4,3, figsize=(15,15))
for i, column in enumerate(col_num):
  sns.histplot(df[column], ax=axes[i//3, i%3], kde=False)
plt.tight_layout() #evitar que se encimen
```

```python
df['job']=df['job'].astype('category')
df['marital']=df['marital'].astype('category')
df['education']=df['education'].astype('category')
df['contact']=df['contact'].astype('category')
df['poutcome']=df['poutcome'].astype('category')
```

```python
df['housing'] = df['housing'].map({'yes': 1, 'no': 0})
df['loan'] = df['loan'].map({'yes': 1, 'no': 0})
df['y']=df['y'].map({'yes':1, 'no':0})
df['month'] = df['month'].map({'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8,
    'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12})
df['day_of_week'] = df['day_of_week'].map({'mon': 1, 'tue': 2, 'wed': 3, 'thu': 4, 'fri': 5})
```

```python
df2=df.copy()
df2
```

```python
df2['duration-log']=np.log(df['duration']+1) #se usa el +1 para evitar indeterminaciones de log(0) en valores cercanos al 0
df2['campaign-log']=np.log(df['campaign']+1)
df2['previous-log']=np.log(df['previous']+1)
```

```python
cols_num=['age','job','marital' ,'education','default','housing','loan','contact','month','day_of_week','duration-log','campaign-log','pdays','previous-log','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed','y']
```

```python
fig, axes=plt.subplots(5,4, figsize=(15,15))
for i, column in enumerate(cols_num):
  sns.histplot(df2[column], ax=axes[i//4, i%4], kde=False)
plt.tight_layout()
```

```python
reporte= df2.select_dtypes(include=["category"])
for col in reporte.columns:
  print(f"Columna: '{col}'")
  print(reporte[col].value_counts())
  print("---------------")
#print()
```

```python
df2['y'] = df2['y'].astype(int)
```

```python
#analisis bivariado
for col in reporte:
    print(f"Tasa de conversiÃ³n (y=1) por {col}:")
    rate = df2.groupby(col)['y'].mean().sort_values(ascending=False)
    print(rate, "\n")
```

```python
fig, axes = plt.subplots(4, 2, figsize=(12, 12))
for i, column in enumerate(reporte):
    row, col = i // 2, i % 2
    rate = df2.groupby(column, observed=True)['y'].mean().sort_values(ascending=False)
    sns.barplot(x=rate.index, y=rate.values, ax=axes[row, col])
    axes[row, col].set_title(f'{column} vs y')
plt.tight_layout()
```

```python
reporte_num= df2.select_dtypes(include=["number"])
for col in reporte_num:
  print(f"Columna: '{col}'")
  print(reporte_num[col].value_counts())
  print("---------------")
```

```python
fig, axes = plt.subplots(8, 2, figsize=(12, 12))
for i, column in enumerate(reporte_num):
    row, col = i // 2, i % 2
    rate = df2.groupby(column, observed=True)['y'].mean().sort_values(ascending=False)
    sns.barplot(x=rate.index, y=rate.values, ax=axes[row, col])
    axes[row, col].set_title(f'{column} vs y')
plt.tight_layout()
plt.show()
```
